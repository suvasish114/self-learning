```python
# imports
import pandas as pd
import statistics as stat
import matplotlib.pyplot as plt
```

## Problem Statement

Given details of 100 different student marks vs study hour dataset; Our objective to draw conclusion of center tendency from sample distribution of the given data.

Dataset: Student Marks Dataset [[kaggle.com](https://www.kaggle.com/datasets/yasserh/student-marks-dataset)]

```python
# reading sample data
df = pd.read_csv('Student_Marks.csv')
df = df.sort_values(by = "time_study")  # sort rows on time_study column
df.head()
```

```python
# extarcting time study vs marks from dataframe
x = list(df["time_study"])
y = list(df["Marks"])
```

```python
# plot
plt.plot(x, y)
plt.vlines(stat.mean(x), ymin=min(y), ymax=max(y), colors='r', linestyles='dashed', label = 'mean')
plt.vlines(stat.median(x), ymin=min(y), ymax=max(y), colors='g', linestyles='dashed', label = 'median')
plt.xlabel("Time Study")
plt.ylabel("Marks")
plt.title("Time Study vs Marks")
plt.legend()
plt.show()
```

## Observation

- Mean is highly sensetive to outliers.
- Extreme high or low values only affect the median if they fall exactly in the middle, which is less likely than them influencing the sum that affects the mean.
